```python
def analyze(f):
    data = np.loadtxt(fname=f, delimiter=',')
    
    fig = plt.figure(figsize=(10.0,3.0))
    
    axes1 = fig.add_subplot(1,3,1)
    axes2 = fig.add_subplot(1,3,2)
    axes3 = fig.add_subplot(1,3,3)
    
    matrix_mean = np.mean(data, axis=0)
    matrix_max = np.max(data, axis=0)
    matrix_min = np.min(data, axis=0)
    
    axes1.set_ylabel('average')
    axes1.set_xlabel('days')
    axes1.plot(matrix_mean)
    
    axes2.set_ylabel('max')
    axes2.set_xlabel('days')
    axes2.plot(matrix_max)
    
    axes3.set_ylabel('min')
    axes3.set_xlabel('days')
    axes3.plot(matrix_min)
    
    plt.show()
    
    return matrix_mean, matrix_max, matrix_min

# if you add "fig" to the return line, it will also save the plot as something you can call back later 
```
